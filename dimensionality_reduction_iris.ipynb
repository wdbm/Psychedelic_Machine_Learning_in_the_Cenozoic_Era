# dimensionality reduction on iris dataset

```python
import matplotlib.pyplot as plt
import matplotlib.ticker
import mpl_toolkits.mplot3d
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets
import sklearn.decomposition
import sklearn.manifold
import sklearn.preprocessing
sns.set(style = 'ticks')
sns.set_palette('husl')
mpl_toolkits.mplot3d.Axes3D
import warnings
warnings.filterwarnings("ignore")
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (17, 17)
#sns.set(context="paper", font="monospace")
```

```python
df = pd.read_csv(
    'iris.csv',
    header = None,
    names  = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
)
df.head()
```

```python
features = list(df.columns[:-1])
print(features)
X = df[features].values
y = df["species"].values
```

```python
df["species"].unique()
```

```python
Y = sklearn.manifold.TSNE(
    n_components = 2,
    init         = "pca",
    method       = "barnes_hut"
).fit_transform(X)
```

```python
for point, label in zip(X, y):
    if label == 'Iris-setosa':
        plt.scatter(point[0], point[1], c="red")
    if label == 'Iris-versicolor':
        plt.scatter(point[0], point[1], c="blue")
    if label == 'Iris-virginica':
        plt.scatter(point[0], point[1], c="green")
    plt.axes().annotate(label, (point[0], point[1]))
```

```python
for point, label in zip(Y, y):
    if label == 'Iris-setosa':
        plt.scatter(point[0], point[1], c="red")
    if label == 'Iris-versicolor':
        plt.scatter(point[0], point[1], c="blue")
    if label == 'Iris-virginica':
        plt.scatter(point[0], point[1], c="green")
    plt.axes().annotate(label, (point[0], point[1]))
```
